### 线性不可分的 SVM

通过将低维空间映射到高维空间，低维不可分的数据可能在高维是可分的。通过高斯核函数，可以直接在低维空间计算出高维空间的点积结果

高斯核函数：
$$
k(x_i, x_j) = exp(-\frac{||x_i-x_j||^2}{2\sigma^2})
$$

$\sigma$ 越大，核函数越陡峭，模型复杂度越高，越容易过拟合； $\sigma$ 越小，模型越简单，越容易欠拟合。

```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
data = sio.loadmat('./data/ex6data2.mat')
data.keys()
```

```python
X, y = data['X'], data['y']
X.shape, y.shape
```

```python
def plot_data():
    plt.scatter(X[:, 0], X[:, 1], c=y.flatten(), cmap='jet')
    plt.xlabel('x1')
    plt.ylabel('x2')
```

```python
plot_data()
```

```python
from sklearn.svm import SVC
```

```python
svc1 = SVC(C=1, kernel='rbf', gamma=1)   # 使用高斯核函数
svc1.fit(X, y.flatten())
```

```python
svc1.score(X, y.flatten())
```

```python
def plot_boundary(model):
    x_min, x_max = 0, 1
    y_min, y_max = 0.4, 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), 
                        np.linspace(y_min, y_max, 500))
    z = model.predict(np.c_[xx.flatten(), yy.flatten()])
    
    zz = z.reshape(xx.shape)
    plt.contour(xx, yy, zz)
```

```python
plot_boundary(svc1)
plot_data()
```

$\sigma$ 比较小，拟合效果不好

```python
svc2 = SVC(C=1, kernel='rbf', gamma=50)
svc2.fit(X, y.flatten())
```

```python
svc2.score(X, y.flatten())
```

```python
plot_boundary(svc2)
plot_data()
```

$\sigma$比较大，拟合效果好

```python
svc3 = SVC(C=1, kernel='rbf', gamma=1000)
svc3.fit(X, y.flatten())
```

```python
svc3.score(X, y.flatten())
```

```python
plot_boundary(svc3)
plot_data()
```

$\sigma$ 非常大，模型复杂度很高，完全拟合
