```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
mat = sio.loadmat('data/ex6data3.mat')
mat.keys()
```

```python
X, y = mat['X'], mat['y']
Xval, yval = mat['Xval'], mat['yval']
```

```python
def plot_data():
    plt.scatter(X[:, 0], X[:, 1], c=y.flatten(), cmap='jet')
    plt.xlabel('x1')
    plt.ylabel('x2')
```

```python
plot_data()
```

```python
Cvalues = [0.01,  0.03, 0.1, 0.3, 1, 3, 10, 30, 100]     # 9
gammavalues = [0.01,  0.03, 0.1, 0.3, 1, 3, 10, 30, 100] # 9
```

```python
best_score = 0
best_parameters = []

for c in Cvalues:
    for gamma in gammavalues:
        svc = SVC(C=c, kernel='rbf', gamma=gamma)
        svc.fit(X, y.flatten())
        score = svc.score(Xval, yval.flatten())
        
        if score == best_score:
            best_parameters.append((c, gamma))
        if score > best_score:
            best_parameters = []
            best_score = score
            best_parameters.append((c, gamma))

print(best_score, best_parameters)
```

有多组参数共同得到了最优解

```python
svc_best = SVC(C=best_parameters[0][0], kernel='rbf', gamma=best_parameters[0][1])
svc_best.fit(X, y.flatten())
```

```python
def plot_boundary(model):
    x_min, x_max = -0.85, 0.5
    y_min, y_max = -0.85, 0.85
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), 
                         np.linspace(y_min, y_max, 500))
    z = model.predict(np.c_[xx.flatten(), yy.flatten()])
    zz = z.reshape(xx.shape)
    plt.contour(xx, yy, zz)
```

```python
plot_boundary(svc_best)
plot_data()
```
